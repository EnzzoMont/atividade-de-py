```python
notas_funcionario1 = [8, 7, 6, -2, 9]
notas_funcionario2 = [6, 5, 7, 8, 7]

def media(listas):
    for i in listas:
        if i < 0:
            listas.remove(i)
    return (lambda x: sum(x) / len(x)) (listas)

print("A média das notas do funcionário 1 foi: ", media(notas_funcionario1))
print("A média das notas do funcionário 2 foi: ", media(notas_funcionario2))
      
```
